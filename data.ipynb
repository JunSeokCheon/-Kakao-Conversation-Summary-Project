1. 2명의 대화 - 20개

```python
conversations = []

paths = glob.glob("D:/297.SNS 데이터 고도화/01-1.정식개방데이터/Validation/02.라벨링데이터/VL/*json")
target_count = 20
count = 0

# conv_dict - header(대화 주체 정보) / body(실제 대화)
for path in paths:
    with open(path, 'r') as f:
        conv_dict = json.load(f)
        # 대화 참가자가 2명인 대화들만 필터링
        if conv_dict['header']['dialogueInfo']['numberOfParticipants'] == 2:
            # 대화의 수가 30개 초과인 대화들만 필터링
            if conv_dict['header']['dialogueInfo']['numberOfUtterances'] == 30:
                conv_list = []
                for d in conv_dict['body']:
                    # 참가자 아이디 + 대화 내역을 conv_list에 append
                    conv_list.append(d['participantID'] + ': ' + d['utterance'])
                
                # 중복 데이터가 존재하여 예외사항 처리
                if conv_list[0] == conv_list[1]:
                    print("Repeated Conversations")
                    continue
                
                conv_text = '\n'.join(conv_list)
                conversations.append(conv_text)
                count += 1
                if count == target_count:
                    break
```

3명 이상의 대화 - 30개

```python
target_count = 20
count = 0

# conv_dict - header(대화 주체 정보) / body(실제 대화)
for path in paths:
    with open(path, 'r') as f:
        conv_dict = json.load(f)
        # 대화 참가자가 3명 이상인 대화들만 필터링
        if conv_dict['header']['dialogueInfo']['numberOfParticipants'] > 2:
            # 대화의 수가 30개 초과인 대화들만 필터링
            if conv_dict['header']['dialogueInfo']['numberOfUtterances'] == 30:
                conv_list = []
                for d in conv_dict['body']:
                    # 참가자 아이디 + 대화 내역을 conv_list에 append
                    conv_list.append(d['participantID'] + ': ' + d['utterance'])
                
                # 중복 데이터가 존재하여 예외사항 처리
                if conv_list[0] == conv_list[1]:
                    print("Repeated Conversations")
                    continue
                
                conv_text = '\n'.join(conv_list)
                conversations.append(conv_text)
                count += 1
                if count == target_count:
                    break
```

3천자 이상의 대화로 변환
- 데이터셋에는 3천자를 넘는 데이터가 단 한개도 존재하지 않음
- Anthropic Claude 3.5 Sonnet을 사용해서 대화를 주고 앞의 내용을 추측한 내용으로 대화 앞에 붙여서 3천자 이상으로 변환

```python
import anthropic
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
# 3명의 대화 1개를 프롬프트로 넣어서 3천자로 변환해달라는 프롬프트 작성
prompt = f"""You are given a conversation among three users below. Imagine what the users discussed prior to the conversation.
Please write the previous content, strictly following the tone and format of given conversation, with at least 3000 characters.

{conversations[-2]}
"""
```

```python
print(conversations[-2])
```

```python
client = anthropic.Anthropic(
    api_key=os.environ['ANTHROPIC_API_KEY']
)
```

```python
message = client.messages.create(
    model="claude-3-5-sonnet-20240620",
    max_tokens=4096,
    temperature=0.0,
    messages=[
        {"role": "user", "content": prompt}
    ]
)

message.usage
```

```python
message.content[0].text.split('\n\n')
```

```python
# llm 모델에 의해서 생성된 텍스트만 저장
lst = message.content[0].text.split('\n\n')[1:-1]
lst
```

```python
import pickle

with open('./res/conv_long.pickle', 'wb') as f:
    pickle.dump(lst, f)
```

```python
# 생성된 내용 + 기존 내용
conversations[-2] = '\n'.join(lst) + '\n' + conversations[-2]
conversations[-2]
```

```python
with open('./res/eval_data.pickle', 'wb') as f:
    pickle.dump(conversations, f)
```
